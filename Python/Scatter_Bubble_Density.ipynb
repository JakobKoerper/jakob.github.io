```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("crimerates-by-state-2005.csv")
data.head()
```

```python
data = data.iloc[1:,]
data.head()
```

```python
plt.scatter(data.robbery, data.murder)
plt.xlabel('Number of Robberies')
plt.ylabel('Number of Murders')
plt.title('Scatter plot 1')
plt.show
```

```python
plt.scatter(data.aggravated_assault, data.murder)
plt.xlabel('Number of Aggravated Assaults')
plt.ylabel('Number of Murders')
plt.title('Scatter plot 2')
plt.show
```

```python
plt.scatter(data.aggravated_assault, data.burglary)
plt.xlabel('Number of Aggravated Assaults')
plt.ylabel('Number of Burglaries')
plt.title('Scatter plot 3')
plt.show
```
